```pyspark
# replace the adl_name <Azuredatalakestore-Name> with the one we created in step 2
adl_loc="adl://<Azuredatalakestore-Name>.azuredatalakestore.net/"

# define input paths for Aggregated Sales Data and Optimization Result Data
Aggregated_Sales_Data_Path=adl_loc+"aggregated_sales_data/"
Optimization_Result_Data_Path=adl_loc+"opt_results_data/"

# Load the parquet files
Aggregated_Sales_Data=sqlContext.read.parquet(Aggregated_Sales_Data_Path)
Optimization_Result_Data=sqlContext.read.parquet(Optimization_Result_Data_Path)

# Register the data as a table to run queries against
Aggregated_Sales_Data.registerTempTable("Aggregated_Sales_Data")
Optimization_Result_Data.registerTempTable("Optimization_Result_Data")
```
