```pyspark
# replace the <AzureBlobStorageName> with the one we created in step 2
blob_storage_name = "<AzureBlobStorageName>"
result_container = "results"

# define input paths for Aggregated Sales Data and Optimization Result Data
Aggregated_Sales_Data_Path = "wasb://{1}@{0}.blob.core.windows.net/".format(blob_storage_name,result_container) + "aggregated_sales_data/"
Optimization_Result_Data_Path="wasb://{1}@{0}.blob.core.windows.net/".format(blob_storage_name,result_container) + "opt_results_data/"

# Load the parquet files
Aggregated_Sales_Data = spark.read.parquet(Aggregated_Sales_Data_Path)
Optimization_Result_Data = spark.read.parquet(Optimization_Result_Data_Path)

# Register the data as a table to run queries against
Aggregated_Sales_Data.registerTempTable("Aggregated_Sales_Data")
Optimization_Result_Data.registerTempTable("Optimization_Result_Data")
```
